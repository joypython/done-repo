```python
import pandas as pd
```

```python
ndf = pd.read_csv('8. Netflix Dataset.csv')
ndf.head(2)
```

```python
# Is there any Duplicate Record in this dataset? If yes, then remove the duplicate records.

ndf.duplicated().sum()
```

```python
ndf.drop_duplicates(inplace=True)
```

```python
# Is there any Null Value present in any column?

ndf.isnull().sum()
```

```python
# For 'House of Cards', what is the Show Id and who is the Director of this show?

filt = ndf['Title']=='House of Cards'
ndf[filt]
```

```python
# In which year highest number of the TV Shows & Movies were released?

ndf['Release_Date'] = pd.to_datetime(ndf['Release_Date'])
```

```python
ndf.head(1)
```

```python
ndf['Year'] = ndf['Release_Date'].dt.strftime('%Y')
```

```python
ndf.head(2)
```

```python
ndf.groupby('Year')['Category'].value_counts().sort_values(ascending=False)
```

```python
# How many Movies & TV Shows are in the dataset?

ndf['Category'].value_counts()
```

```python
# Show all the Movies that were released in year 2020.

ndf[(ndf['Category'] == 'Movie') & (ndf['Year'] == "2020") ]
```

```python
# Show only the Titles of all TV Shows that were released in India only.

filt = (ndf['Country'] =='India') & (ndf['Category'] =='TV Show')
ndf[filt]['Title']
```

```python
# Show Top 10 Directors, who gave the highest number of TV Shows & Movies to Netflix?

ndf.groupby('Director')['Category'].value_counts().nlargest(10)
```

```python
# Show all the Records, where "Category is Movie and Type is Comedies" or "Country is United Kingdom".

filt = (ndf['Category'] == 'Movie') & (ndf['Type'] == 'Comedies') | (ndf['Country'] == 'United Kingdom')
ndf[filt]
```

```python
# In how many movies/shows, Tom Cruise was cast?

filt1 = ndf['Cast'].str.contains('Tom Cruise')==True
ndf[filt1]
```

```python
# What are the different Ratings defined by Netflix?

ndf['Rating'].unique()
```

```python
# How many Movies got the 'TV-14' rating, in Canada?

filt = (ndf['Rating'] == 'TV-14') & (ndf['Country'] == 'Canada') & (ndf['Category'] == 'Movie')
len(ndf[filt])
```

```python
ndf['Year'] = ndf['Year'].astype('float')
```

```python
# How many TV Show got the 'R' rating, after year 2018?

filt = (ndf['Category'] =='TV Show') & (ndf['Rating'] =='R') & (ndf['Year'] > 2018)
ndf[filt]
```

```python
# Which individual country has the Highest No. of TV Shows?

filt = ndf['Category'] == 'TV Show'
ndf[filt][['Category','Country']].value_counts().nlargest(1)
```

```python
# How can we sort the dataset by Year?

ndf.sort_values('Year', ascending=False, inplace=True)
ndf.head()
```

```python
# Find all the instances where: Category is 'Movie' and Type is 'Dramas' or Category is 'TV Show' & Type is 'Kids' TV'.

filt = (ndf['Category'] == 'Movie')&(ndf['Type'] == 'Dramas')|(ndf['Category'] == 'TV Show')&(ndf['Type'] == "Kids' TV")
ndf[filt]
```
