```python
cdf = pd.read_csv('Cars Data1.csv')
cdf.head(2)
```

```python
cdf.isnull().sum()
```

```python
m = cdf['Cylinders'].mean()
m
```

```python
cdf['Cylinders'].fillna(m, inplace=True)
```

```python
cdf['Make'].unique()
```

```python
cdf['Make'].value_counts()
```

```python
filt =(cdf['Origin']=='Asia') | (cdf['Origin']=='Europe')
cdf[filt]
```

```python
filt = cdf['Weight'] > 4000
cdf = cdf[~filt]
cdf
```

```python
cdf['MPG_City'] = cdf['MPG_City'].apply(lambda x: x+3)
cdf
```

```python
cdf['Make'] = cdf['Make'].apply(lambda x: x.upper())
cdf
```
