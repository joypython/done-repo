```python
idf = pd.read_csv('6. India Census 2011.csv')
idf.head()
```

### Calculate the Percentage of Population of Male and Female Voters in each State

```python
mf_pct3 = idf.groupby('State_name')[['Male','Female']].sum() / idf[['Male','Female']].sum()*100
mf_pct3
```

### Calculate Percentage of Population with Different Religion State Wise

```python
pct_rel = idf.groupby('State_name')[['Hindus','Muslims','Christians','Sikhs','Buddhists','Jains']].sum()
pct_rel
```

```python
pct = pct_rel / pct_rel[['Hindus','Muslims','Christians','Sikhs','Buddhists','Jains']].sum()*100
pct
```

### Display Top 10 States based Religion Jains

```python
top_10 = idf.groupby('State_name')['Jains'].sum().nlargest(10)
top_10
```
